```python
X = df.drop(['medv'], axis=1)
y = df['medv']
```

```python
print(X)
print(y)
```

Spliting data into training and testing

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

Evaluation

```python
prediction = model.predict(X_train)
```

```python
print(prediction)
```

```python
#R square error
score_1 = metrics.r2_score(y_train, prediction)

#Mean Absoulte error
score_2 = metrics.mean_absolute_error(y_train, prediction)

print("R square Error :", score_1)
print("Mean Absoulte Error :", score_2)
```

```python
plt.scatter(y_train, prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
test_prediction = model.predict(X_test)
```

```python
print(test_prediction)
```

```python
score_3 = metrics.r2_score(y_test, test_prediction)
score_4 = metrics.mean_absolute_error(y_test, test_prediction)

print("R square Error :", score_3)
print("Mean Absoulte Error :", score_4)
```
